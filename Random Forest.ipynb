```python
import pandas as pd
train = pd.read_csv('1dataset_train.csv', header = 0,)
b = []
for i in train['Data']:
    b.append([int(j) for j in i.lstrip('[').rstrip(']').split(', ')])
```

```python
train
```

```python
from sklearn.model_selection import train_test_split
X = b
y = [int(i) for i in train['Class_label_FPG']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=85)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(y_test)
```

```python
k = 0
g = 0
c = [int(i) for i in train['Class_label_FPG']]
a = clf.predict(b)
for i in range(len(a)):
    if(c[i] == 1):
        g += 1
    else:
        g += 2
    k += abs(a[i] - c[i])
print(1 - (k / g))
```

```python
k = 0
c = [int(i) for i in train['Class_label_FPG']]
a = clf.predict(b)
for i in range(len(a)):
    if a[i] != c[i]:
        k += 1
print(k / len(a))
```
